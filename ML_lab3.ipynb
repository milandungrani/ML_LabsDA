```python
exf = pd.read_csv (r'/home/user/Downloads/data1.csv')
y=exf.y
x1=(exf.x1-exf.x1.mean())/exf.x1.std()
x2=(exf.x2-exf.x2.mean())/exf.x2.std()

y=y[0:90]
x1=x1[0:90]
x2=x2[0:90]
def sigmoid(x):
    return 1.0/(1 + np.exp(-x))

curr1=np.random.randint(10)+1
curr2=np.random.randint(10)+1
curr3=np.random.randint(10)+1
rate=0.01
lim=0.0000001
diff1=1
diff2=1
diff3=1
i=0
n=len(x1)
def df1(curr1,curr2,curr3):
    err=0
    for it in range(0,n):
        err+=(sigmoid(curr1+(curr2)*x1[it]+(curr3)*x2[it])-y[it])
    return err

def df2(curr1,curr2,curr3):
    err=0
    for it in range(0,n):
        err+=(sigmoid(curr1+(curr2)*x1[it]+(curr3)*x2[it])-y[it])*x1[it]
    return err

def df3(curr1,curr2,curr3):
    err=0
    for it in range(0,n):
        err+=(sigmoid(curr1+(curr2)*x1[it]+(curr3)*x2[it])-y[it])*x2[it]
    return err


while diff1>lim and diff2>lim and diff3>lim and i<10000:
    prev1=curr1
    prev2=curr2
    prev3=curr3
    curr1=curr1-rate*(df1(prev1,prev2,prev3))
    curr2=curr2-rate*(df2(prev1,prev2,prev3))
    curr3=curr3-rate*(df3(prev1,prev2,prev3))
    diff1=abs(curr1-prev1)
    diff2=abs(curr2-prev2)
    diff3=abs(curr3-prev3)
    i+=1
print(curr1,curr2,curr3)
y_test=exf.y[0:100]
x1_test=(exf.x1-exf.x1.mean())/exf.x1.std()
x2_test=(exf.x2-exf.x2.mean())/exf.x2.std()
error=0
for i in range(0,100):
    if((sigmoid(curr1+(curr2*x1_test[i])+(curr3*x2_test[i]))>=0.5 and y_test[i]==0) or (sigmoid(curr1+(curr2*x1_test[i])+(curr3*x2_test[i]))<0.5 and y_test[i]==1)):
        error+=1
        #print(i)
print("Error - ",error)
print("Accuracy - ",(len(y_test)-error)*100/len(y_test))
z=(-curr1-(curr2*x1))/curr3
for i in range(0,len(x1)):
    if y[i]==0:
        plt.scatter(x1[i],x2[i],color='red')
    else:
        plt.scatter(x1[i],x2[i],color='blue')
plt.plot(x1,z)
plt.show()
```
